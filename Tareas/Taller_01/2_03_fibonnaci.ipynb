```python
import matplotlib.pyplot as plt
import numpy as np
```

# Fibonnaci

```python
def fibo_gen(n=None):
    a, b = 0, 1
    counter = 0
    while not n or counter < n:
        if counter == 0:
            counter += 1
            yield a
        elif counter == 1:
            counter += 1
            yield b
        else:
            current = a + b
            a, b = b, current
            counter += 1
            yield current
```

```python
fibonacci = list(fibo_gen(20))
```

```python
plt.plot(fibonacci, 'r--', markersize=10)
plt.show()
```

# Golden ratio

```python
phi = fibonacci[19]/fibonacci[18]
phi_teo = (1+np.sqrt(5))/2
print(f'Fibonnaci teorico: {phi_teo}')
print(f'Fibonnaci calculado: {phi}')
print(f'Error: {phi_teo-phi}')
```

```python
phi_list = [fibonacci[i]/fibonacci[i-1] for i in range(2, len(fibonacci))]
plt.plot(phi_list-phi_teo, 'b--', markersize=10)
plt.show()
```
