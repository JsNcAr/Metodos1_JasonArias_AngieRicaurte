```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import urllib.request
```

```python
url="https://raw.githubusercontent.com/asegura4488/Database/main/MetodosComputacionalesReforma/EstrellaEspectro.txt"
filename="DatosMaximo.txt"
urllib.request.urlretrieve(url,filename)
data=np.loadtxt(filename)
plt.plot(data[:,0],data[:,1])
```

```python
from operator import length_hint
datos=np.array(data)

def maximos_locales(array):
  maximos=[]
  for i in range(1,len(datos)-1):
    anterior=datos[i-1][1]
    actual_x=datos[i][0]
    actual=datos[i][1]
    siguiente=datos[i+1][1]
    if actual>anterior and actual>siguiente:
      maximos.append([actual_x,actual])
  return maximos
maximos_l=maximos_locales(datos)

x=[]
y=[]
for i in maximos_l:
  coordenada_x=i[0]
  x.append(coordenada_x)
  coordenada_y=i[1]
  y.append(coordenada_y)
plt.scatter(x,y,c="red")
plt.plot(data[:,0],data[:,1])
plt.show()
  



    
    




    





  

 
```
