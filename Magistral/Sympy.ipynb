```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import sympy as sym
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = sym.Symbol('x')
y = sym.Symbol('y')
```

```python
_x = np.linspace(0, 2*np.pi, 100)
_y = np.linspace(0, 2*np.pi, 100)
```

```python
F = sym.sin(3*x)*sym.exp(-0.5*x)
F
```

```python
F = sym.lambdify([x], F, 'numpy')
```

```python
plt.plot(_x, F(_x))
```

```python
F2 = sym.exp(-x**2 - y**2)
F2
```

```python
DxF2 = sym.diff(F2, x, 2)
DxF2
```

```python
sym.integrate(F2, (x, -sym.oo, sym.oo), (y, -sym.oo, sym.oo))
```

```python
Grid = sym.lambdify([x, y], F2, 'numpy')
_x = np.linspace(-2, 2, 100)
_y = np.linspace(-2, 2, 100)

X, Y = np.meshgrid(_x, _y)
Z = Grid(X, Y)
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(121, projection='3d')
ax1 = fig.add_subplot(122)
ax.plot_surface(X, Y, Z, cmap='viridis')
ax1.contour(X, Y, Z, cmap='viridis')
plt.show()
```
